```python
# Creating a copy of the original data frame
df_copy = df.copy()
```

```python
# Dropping the data
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
df_copy.shape
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
# Replacing M with 000
df_copy['Size'] = df_copy['Size'].str.replace('M', "000")
```

```python
# Replacing k with 000
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
```

```python
df_copy['Size'].unique()
```

```python
# Replacing Null Values with NAN
df_copy['Size'] = df_copy['Size'].str.replace('Varies with device', str(np.nan))
```

```python
df_copy['Size'].unique()
```

```python
# Changing the data type of Size column
df_copy['Size'] = df_copy['Size'].astype("float")
```

```python
df_copy['Size'].unique()
```

```python
# Replacig the decimal values by multiplying with 1000 where the vlaue is less than 10.
for i in df_copy['Size']:
    if i<10:
        df_copy['Size'] = df_copy['Size'].replace(i,i*1000)
    
```

```python
df_copy['Size'].unique()
```

```python
# Scaling down the values of the size column 
df_copy['Size'] = df_copy['Size']/1000
```

```python
df_copy['Size'].unique()
```

```python
# Data type of the Installs column
df_copy['Installs'].dtype
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
# Removing the characters from the two columns Installs and Price

char_to_remove = ['+', ',', '$']
cols_to_clean = ['Installs', 'Price']

for items in char_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(items, '')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
# Changing the data type of the Install to Int
df_copy['Installs'] = df_copy['Installs'].astype('int')
```

```python
# Changing the data type of the Price to Int
df_copy['Price'] = df_copy['Price'].astype('float')
```

```python
# Converting Last Updated to date time format
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
```

```python
# Creating a new column with Day based on the Last Updated column
df_copy['Day'] = df_copy['Last Updated'].dt.day
```

```python
# Creating a new column with Month based on the Last Updated column
df_copy['Month'] = df_copy['Last Updated'].dt.month
```

```python
# Creating a new column with Year based on the Last Updated column
df_copy['Year'] = df_copy['Last Updated'].dt.year
```
